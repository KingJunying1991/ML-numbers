```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 读取文件
train = []
target=[]
for i in range(10):
    for j in range(1,501):
        img_path = './digits/%d/%d_%d.bmp'%(i,i,j)
        # 做一个扁平化,才可以机器训练
        img_data = plt.imread(img_path).ravel()
        train.append(img_data)
        # 每获取到一个图片数据，就把对应的target保存下来
        target.append(i)
```

```python
plt.imshow(plt.imread('./digits/0/0_1.bmp'))
```

```python
# 数据切分
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(train,target)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knnclf = KNeighborsClassifier(n_neighbors=5)
knnclf.fit(X_train,y_train)
```

```python
knnclf.score(X_test,y_test)
```

```python
y_ = knnclf.predict(X_test)
```

```python
# 绘制图形(预测值和真实值表达出来)
plt.figure(figsize=(16,16))
for i in range(100):
    axes = plt.subplot(10,10,i+1)
    plt.imshow(X_test[i].reshape(28,28),cmap='gray')
    t = y_test[i]
    p = y_[i]
    title = 'T:'+str(t)+'\nP:'+str(p)
    axes.set_title(title)
    axes.axis('off')
```
